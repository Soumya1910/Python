```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv('data/gas_price.csv')
```

```python
gas.head()
```

```python
country = list(gas.columns)[1:]
country
```

```python
plt.figure(figsize=(10,6))
plt.title('Gas Price in USD Dollar', fontdict={'fontsize':18, 'fontweight':'bold'})

plt.plot(gas.Year, gas['USA'], label='United States', marker='.')
plt.plot(gas.Year, gas['UK'], label='United Kingdom', marker = 'o')
plt.plot(gas.Year, gas['Canada'], 'g+-', label='United Kingdom')


year_list = gas['Year'][::2]
#plt.xticks(gas.Year)
plt.xticks(year_list)


plt.xlabel('Year')
plt.ylabel('Price')


plt.legend()
plt.show()
```
